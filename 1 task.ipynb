```python
import os
import pandas as pd 
import numpy as np
from io import StringIO 

def answer_one():
    df = pd.ExcelFile('Energy Indicators.xls')
    energy = df.parse(skiprows=17,skip_footer=(38))
    energy.drop(energy.columns[[0,1]], axis=1, inplace=True) 
    energy.columns=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    yulo=energy['Energy Supply']
    mas=yulo.values
    a=[]
    i=0
    for i in mas:
        if i=="Petajoules":
            energy.loc[i,('Energy Supply')]= 1000000
        if i=="...":
            energy.loc[i,('Energy Supply')]= np.NaN
    i=0
    mafa=energy['Country']
    for i in mafa:
        j=str(i) 
        o=list(j)
        if o!="nan":
            for word in o:
                a.append(word)
                my = ''.join(map(str, a))
                if word=="(":
                    a.remove(word)
                    a.pop(-1)
                    my = ''.join(map(str, a))
                    energy.loc[i,('Country')]= my
                    break
                po=str(word) 
                if po.isdigit():
                    a.remove(word)
                    my = ''.join(map(str, a))
                    energy.loc[i,('Country')]= my
                    break
                    
            pot=0
            del a[:]

    i=0  
    masa=energy['Country']
    for i in masa:
        if i=="United States of America":
            energy.loc[i,('Country')]= "United States"
        if i=="Republic of Korea":
            energy.loc[i,('Country')]= "South Korea"
        if i=="United Kingdom of Great Britain and Northern Ireland":
            energy.loc[i,('Country')]= "United Kingdom"
        if i=="China, Hong Kong Special Administrative Region":
            energy.loc[i,('Country')]= "Hong Kong" 
    
    
    GDP= pd.read_csv("GDP.csv",skiprows=4)
    masa1=GDP['Country Name']
    i=0
    for i in masa1:
        if i=="Korea, Rep.":
            GDP.loc[i,('Country Name')]= "South Korea"
        if i=="Iran, Islamic Rep.":       
            GDP.loc[i,('Country Name')]= "Iran"
        if i=="Hong Kong SAR, China":
            GDP.loc[i,('Country Name')]= "Hong Kong"
    GDP=GDP.loc[:, ['Country Name','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']]
    GDP.columns = ['Country','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']
    ScimEn = pd.read_excel(io='scimagojr.xlsx')
    ScimEn = ScimEn[:15]
    df6 = pd.merge(ScimEn,energy,how='inner',left_on='Country',right_on='Country')
    new_df = pd.merge(df6,GDP,how='inner',left_on='Country',right_on='Country')
    new_df = new_df.set_index('Country')
    return new_df
answer_one()

```
