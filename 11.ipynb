```python
 def tratata(kernel,n,fd=load_digits()):
    digits =  fd
    X, y, labels = digits.data, digits.target, digits.target_names
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
    # y = y == 1
    clf = SVC(kernel = kernel).fit(X_train, y_train)
    y_predicted = clf.predict(X_test)
#     print(cross_val_score(clf, X, y, cv=3))
    confusion = confusion_matrix(y_test, y_predicted)
    df = pd.DataFrame(confusion,index = np.arange(n), columns = np.arange(n))
    # clf = SVC()
    # clf = MLPClassifier()
    plt.figure(figsize=(10,5))
    sns.heatmap(df, annot=True, cmap='Reds') #  "YlGnBu")
    plt.title('SVM {0} Kernel \nAccuracy:{1:.3f}'.format(kernel,accuracy_score(y_test,y_predicted)))
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    print(classification_report(y_test, y_predicted))
#     print('Cross-validation (accuracy)', cross_val_score(clf, X, y, cv=5))
#     print('Cross-validation (AUC)', cross_val_score(clf, X, y, cv=5, scoring = 'roc_auc'))
#     print('Cross-validation (recall)', cross_val_score(clf, X, y, cv=5, scoring = 'recall'))
    
```
